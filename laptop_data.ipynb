```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df =  pd.read_csv("laptop_data.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.insert(8,"other",["0" for x in range(len(df['Memory']))])
df.insert(9,"SSD",["0" for x in range(len(df['Memory']))])
df.insert(10,"HDD",["0" for x in range(len(df['Memory']))])
df.insert(11,"Hybrid",["0" for x in range(len(df['Memory']))])
df.insert(12,"Flash Storage",["0" for x in range(len(df['Memory']))])
```

```python
df.head()
```

```python
df['Memory'].unique()
```

```python
for i in range(len(df['Memory'])):
    arr = df.loc[i,'Memory'].split(" +  ")
    df.loc[i,'Memory'] = arr[0]
    if len(arr)>1 :
        df.loc[i,'other'] = arr[1]
```

```python
df.head()
```

```python
for i in range(len(df['Memory'])):
    
    arr = df.loc[i,'Memory'].split()
    
    if "GB" in arr[0]:
        final = arr[0].replace("GB"," ")
    else: 
        final = float(arr[0].replace("TB"," "))*1024

    
    if "SSD" in df.loc[i,'Memory']:
        df.loc[i,'SSD'] = final
    elif "HDD" in df.loc[i,'Memory']:
        df.loc[i,'HDD'] = final
    elif "Hybrid" in df.loc[i,'Memory']:
        df.loc[i, 'Hybrid'] = final
    else:
        df.loc[i,"Flash Storage"] = final           
```

```python
df['other'].unique()
```

```python
df.head()
```

```python
df['other'].unique()
```

```python
df['SSD'] = df['SSD'].astype(int)
df['HDD'] = df['HDD'].astype(int)
df['Hybrid'] = df['Hybrid'].astype(int)

for i in range(len(df['Memory'])):
    array = df.loc[i, 'other'].split()

    if "GB" in array[0]:
        final = int(float(array[0].replace("GB"," ")))
    else: 
        final = int(float(array[0].replace("TB"," ")))*1024
        
    if "SSD" in df.loc[i,'other']:
        df.loc[i,'SSD'] += final
    elif "HDD" in df.loc[i,'other']:
        df.loc[i,'HDD'] += final
    elif "Hybrid" in df.loc[i,'other']:
        df.loc[i, 'Hybrid'] += final  
```

```python
df.head(3)
```

```python
df = df.drop(columns = ["Memory","other"])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['ScreenResolution'].unique()
```

```python
df.insert(5,"IPS",["0" for x in range(len(df['ScreenResolution']))])
df.insert(6,"Touchscreen",["0" for x in range(len(df['ScreenResolution']))])

df['IPS'] = df['ScreenResolution'].str.contains('IPS').astype(int)
df['Touchscreen'] = df['ScreenResolution'].str.contains('Touchscreen').astype(int)
df.head()
```

```python
df['Resolution'] = df['ScreenResolution'].str.extract(r'(\d+x\d+)')
df.insert(7,"width",["0" for x in range(len(df['ScreenResolution']))])
df.insert(8,"height",["0" for x in range(len(df['ScreenResolution']))])
for i in range(len(df['Resolution'])):
    arr = df.loc[i,'Resolution'].split("x")
    df.loc[i,'width'] = arr[0]
    df.loc[i,'height'] = arr[1]
    
df = df.drop(columns=["ScreenResolution",'Resolution'])
df.head()
```

```python
df['Ram'] = df['Ram'].str.replace("GB","").str.strip()
```

```python
df.head()
```

```python
df['Cpu'].head()
```

```python
df.insert(9,"Gen",["0" for x in range(len(df['Cpu']))])
for i in range(len(df['Cpu'])):
    
    arr =  df.loc[i,'Cpu'].split()
    length = len(arr)-1
    df.loc[i,'Gen'] = arr[length]
df.head()
df['Cpu'] = df['Cpu'].apply(lambda x : " ".join(x.split()[:3]))
df['Gen'] = df['Gen'].str.replace("GHz","")
df.head()
```

```python
df.insert(11,"Gpu Brand",["0" for x in range(len(df['Gpu']))])
df['Gpu Brand'] = df['Gpu'].apply(lambda n:n.split()[0])
df['Gpu Brand']
```

```python
df['Gpu'] = df['Gpu'].apply(lambda x:" ".join(x.split()[1:]))
df.rename(columns = {"Gpu":'Gpu Name'},inplace=True)
df.head()
```

```python
df['Weight'] = df['Weight'].str.replace("kg","")
```

```python
df.head()
```

```python
df['Weight'].max()
```

```python
df['Company'].unique()
```

```python
df['TypeName'].unique()
```

```python
enc_cols = ['Company','TypeName','Cpu','Gpu Brand','Gpu Name','OpSys']
encoders = {}
```

```python
for col in enc_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le
```

```python
df.head()
```

```python
df.info()
```

```python
cols_convert = ['width', 'height','Gen','Ram','Flash Storage','Weight']
df[cols_convert] = df[cols_convert].apply(pd.to_numeric)
```

```python
df.info()
```

```python
df['HDD'].unique()
```

```python
X = df[['Company', 'TypeName', 'Inches', 'IPS', 'Touchscreen',
       'width', 'height', 'Cpu', 'Gen', 'Ram', 'Gpu Brand', 'SSD', 'HDD',
       'Hybrid', 'Flash Storage', 'Gpu Name', 'OpSys', 'Weight']]
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 51)
```

```python
X_train.shape, X_test.shape,
y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(X_train, y_train)
```

```python
linear_model.intercept_
```

```python
linear_model.coef_
```

```python
y_pred_test = linear_model.predict(X_test)
# y_pred_test
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
```

```python
accuracy = r2_score(y_test,y_pred_test)
accuracy
```

```python
rl = Ridge()
rl.fit(X_train,y_train)
rl.score(X_test, y_test)
```

```python
rl2 = Ridge(alpha = 2)
rl2.fit(X_train,y_train)
rl2.score(X_test, y_test)
```

```python
ridge_model = Ridge(alpha = 3)
ridge_model.fit(X_train,y_train)
ridge_model.score(X_test, y_test)
```

```python
lr = Lasso()
lr.fit(X_train,y_train)
lr.score(X_test, y_test)
```

```python
lr2 = Lasso(alpha=2)
lr2.fit(X_train,y_train)
lr2.score(X_test, y_test)
```

```python
lasso_model = Lasso(alpha=7)
lasso_model.fit(X_train,y_train)
lasso_model.score(X_test, y_test)
```
